```python
import pandas as pd

# Load dataset
df = pd.read_csv('epl-goalScorer(20-21).csv')

# 1. 10 Player terbanyak mencetak gol
top_scorers = df.nlargest(10, 'goals')
print("1. 10 Player terbanyak mencetak gol:")
print(top_scorers)

# 2. 10 player terlama durasi bermain
top_minutes_players = df.nlargest(10, 'time')
print("2. 10 Player dengan durasi bermain terlama:")
print(top_minutes_players[['player_name', 'time']])

# 3. 10 player terbanyak assist
top_assists = df.nlargest(10, 'assists')
print("\n3. 10 player terbanyak assist:")
print(top_assists[['player_name', 'assists']])

# 4. 4 player pencetak gol terbanyak dari kesebelasan Manchester City
mc_top_scorers = df[df['team_title'] == 'Manchester City'].nlargest(4, 'goals')
print("\n 4. 4 player pencetak gol terbanyak dari Manchester City:")
print(mc_top_scorers[['player_name', 'goals']])

# 5. Dari kesebelasan mana pencetak gol terbanyak
most_goals_squad = df.groupby('team_title')['goals'].sum().idxmax()
print("\n 5. Pencetak gol terbanyak berasal dari kesebelasan:", most_goals_squad)

# 6. Jumlah kartu kuning dan merah yang diterima oleh pembuat assist terbanyak
top_assist_player = df.nlargest(1, 'assists')['player_name'].values[0]
cards_top_assist = df[df['player_name'] == top_assist_player][['yellow_cards', 'red_cards']].sum()
print("\n 6. Jumlah kartu kuning dan merah yang diterima oleh", top_assist_player, ":")
print(cards_top_assist)

#7. Pemain dengan jumlah assist dan gol terbanyak dan dari kesebelasan mana
combined_stats = df.assign(combined_goals_assists=df['goals'] + df['assists'])
top_combined_player = combined_stats.nlargest(1, 'combined_goals_assists')
print("\n 7. Pemain dengan jumlah assist dan gol terbanyak:", top_combined_player[['player_name', 'combined_goals_assists', 'team_title']].values)

# 8. 5 kesebelasan yang menyumbangkan gol terbanyak
top_goal_contributors = df.groupby('team_title')['goals'].sum().nlargest(5)
print("\n 8. 5 kesebelasan yang menyumbangkan gol terbanyak:")
print(top_goal_contributors)

# 9. Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu kuning atau merah
fair_play_players = df[(df['yellow_cards'] == 0) & (df['red_cards'] == 0)].nlargest(10, 'goals')
print("\n 9. Pemain dengan jumlah gol terbanyak tanpa mendapatkan kartu kuning atau merah:")
print(fair_play_players[['player_name','goals']])
```
